```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
#distrubution of legit transactions and faudenlent transaction
credit_card_data['Class'].value_counts()
```

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class ==0]
fraud = credit_card_data[credit_card_data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure of the  data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing the value of both transaction
credit_card_data.groupby('Class').mean()
```

## Number of fraudulent Transaction

```python
legit_sample = legit.sample(n=338)
```

## concatenating two dataframes

```python
new_dataset =pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## Splitting the data into features and targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Split the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test =train_test_split(X, Y, test_size=0.2, stratify=Y)
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

## Model training

Logistic Regression

```python
model =LogisticRegression()
```

```python
#traing the logistic regression mode with traing data
model.fit(X_train,Y_train)
```

```python
model = LogisticRegression(solver='lbfgs',max_iter=10000)
```

```python
model.fit(X_train,Y_train)
```

## Model Evaluation

Accuracy score

```python
#accuracy on training dat
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print('accracy on training data:',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy =accuracy_score(X_test_prediction , Y_test)
```

```python
print('Accuracy score on test Data : ',test_data_accuracy)
```
